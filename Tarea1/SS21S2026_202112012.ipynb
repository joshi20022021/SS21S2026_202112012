```python
import pandas as pd

# Ruta al archivo en drive
file_path = '/content/drive/MyDrive/dataset_sucio.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# dataset antes de la transformacion
df.head()
```

```python
# Eliminar duplicados
df = df.drop_duplicates()

# Eliminar filas si faltan datos esenciales
df = df.dropna(subset=['id_cliente', 'nombre', 'genero'])

# conversion a string, y reemplazo de comas por puntos y transformar a número
df['gasto_q'] = df['gasto_q'].astype(str).str.replace(',', '.', regex=False)
df['gasto_q'] = pd.to_numeric(df['gasto_q'], errors='coerce')

# Rellenar vacíos con 0 y redondear a 2 decimales
df['gasto_q'] = df['gasto_q'].fillna(0)
df['gasto_q'] = df['gasto_q'].round(2)

# Rellenar vacíos en texto
df['ciudad'] = df['ciudad'].fillna('Desconocido')

# Reformatear fecha al formato correcto
df['fecha_registro'] = pd.to_datetime(df['fecha_registro'], errors='coerce', format='mixed', dayfirst=True)
df['fecha_registro'] = df['fecha_registro'].dt.strftime('%Y-%m-%d')

# Estandarizacion de mayusculas y minusculas
df['nombre']    = df['nombre'].str.strip().str.title()
df['ciudad']    = df['ciudad'].str.strip().str.title()
df['categoria'] = df['categoria'].str.strip().str.lower()
df['genero']    = df['genero'].str.strip().str.lower()

df.head()
```
